```python
def plotMe(dig, label=None):
    if len(dig.shape)==1:
        dig=dig.reshape(28,28)
    plt.imshow(dig,cmap='gray')
    plt.xticks(())
    plt.yticks(())
    plt.xlabel(label)
```

```python
def plotSamples(data):
    plt.figure(figsize=(10,7))
    plt.title("Samples")
    fig=1
    for i in range(data.values.shape[0]):
        if i%10000==0:
            plt.subplot(2,3,fig)
            if data.shape[1]==785:
                plotMe(data.iloc[i,1:].values,str(data.iloc[i,0]))
            else:
                plotMe(data.iloc[i,:].values,)
            fig=fig+1
    plt.show()
```

```python
print("Training Data:")
plotSamples(data)
print("Submission Data:")
plotSamples(data_sub)
```

```python
data_target=data['label']
def plotSampleCounts():    
    ex=data_target.value_counts().index
    why=data_target.value_counts().values
    plt.figure(figsize=(7,5))
    plt.title("Value Counts")
    plt.bar(ex,why)
    plt.ylim(3500,5000)
    plt.xlabel("Digits")
    plt.ylabel("No. of Samples")
    for x,y in zip(ex,why):
        plt.text(x-0.4,y+15,y)
    plt.xticks(ex)
    plt.show()
```

## Performance Function

```python
def plotPerformance(model):
    eval_params=model.history.history
    print(eval_params)
    LossAccPlot=plt.figure(figsize=(20,5))
    plt.title("Loss and Accuracy Vs Epochs")
    plt.subplot(1,2,1)
    plt.plot(eval_params['loss'],label='Loss')
    plt.plot(eval_params['val_loss'],'--',label='Validation Loss')
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.legend()
    plt.subplot(1,2,2)
    plt.plot(eval_params['acc'],label='Accuracy')
    plt.plot(eval_params['val_acc'],'--',label='Validation Accuracy')
    plt.xlabel("Epochs")
    plt.ylabel("Accuracy")
    plt.legend()
    plt.show()
    plt.close()
```

## CNN

```python
def compareModels(epochs):
    comparison_params={
        'train_acc':[],
        'test_acc':[],
        'train_loss':[],
        'test_loss':[]
    }
    
    for epoch in epochs:
        model= Sequential()

        model.add(Conv2D(64,(3,3), input_shape=X.shape[1:]))
        model.add(Activation("relu"))
        model.add(MaxPooling2D(pool_size=(2,2)))

        model.add(Conv2D(128,(3,3)))
        model.add(Activation("relu"))
        model.add(MaxPooling2D(pool_size=(2,2)))

        model.add(Flatten())
        model.add(Dense(64))

        model.add(Dense(10))
        model.add(Activation('sigmoid'))
        print("=================================================================\nModel with", epoch,"epochs:")
        model.compile(loss='categorical_crossentropy', optimizer='sgd', metrics=['accuracy'])
        model.fit(X_train,y_train,batch_size=10, validation_split=0.1, epochs=epoch)
        
        plotPerformance(model)
        
        predicted=np.argmax(model.predict(X_test), axis=1)
        real=np.argmax(y_test,axis=1)
        
        train_acc=model.evaluate(X_train,y_train)[1]
        test_acc=model.evaluate(X_test,y_test)[1]
        train_loss=model.evaluate(X_train,y_train)[0]
        test_loss=model.evaluate(X_test,y_test)[0]
        comparison_params['train_acc'].append(train_acc)
        comparison_params['test_acc'].append(test_acc)
        comparison_params['train_loss'].append(train_loss)
        comparison_params['test_loss'].append(test_loss)
        
        print("Accuracy on Train Data:", train_acc,"\nAccuracy on Test Data:", test_acc,'\n=================================================================')
    
    plt.figure(figsize=(20,10))
    plt.title("Comparison of Models")
    plt.plot(comparison_params['train_acc'], label="Train Accuracy")
    plt.plot(comparison_params['test_acc'], label="Test Accuracy")
    plt.plot(comparison_params['train_loss'],'--', label="Train Loss")
    plt.plot(comparison_params['test_loss'],'--', label="Test Loss")
    plt.xticks(range(len(epochs)),epochs)
    plt.xlabel("Max Epochs")
    plt.ylabel("Accuracy")
    plt.legend()
    plt.show()
```

```python
compareModels([2,4,6,8,10])
```

### 1. Image Translation

```python
def cropImg(img,size=28,direction=None):
    if direction=='tl':
        img=img[-size:,-size:]
    elif direction=='tr':
        img=img[-size:,:size]
    elif direction=='bl':
        img=img[:size,-size:]
    elif direction=='br':
        img=img[:size,:size]
    img=cv2.resize(img,(28,28))
    return img
```

```python
plt.figure(figsize=(5,5))
for i,direc in zip(range(4),['tl','tr','bl','br']):
    plt.subplot(2,2,i+1)
    plt.imshow(cropImg(-X_test[0],20,direc),cmap='gray')
    plt.xticks(())
    plt.yticks(())
plt.show()
```

### 2. Image Flip

```python
def imgFlip(img, axis):
    img=img.reshape(28,28)
    return np.flip(img, axis=axis)
```

```python
plt.figure(figsize=(5,2.5))
for i, axis in zip(range(2), [0,1]):
    plt.subplot(1,2,i+1)
    plt.imshow(imgFlip(-X_test[0],axis),cmap='gray')
    plt.xticks(())
    plt.yticks(())
plt.show()
```

### 3. Image Rotate

```python
def rotateImg(img, deg):
    Imgrot=np.zeros((28,28), dtype=float)
        
    ##Rotation Matrix
    theta = np.radians(deg)
    c, s = np.cos(theta), np.sin(theta)
    R = np.array(((c, -s), (s, c)))

    for i in range(28):
        for j in range(28):
            val=(img[i][j])
            coord=np.dot(R,[i,j])
            if coord[0]<28 and coord[1]<28:
                Imgrot[int(coord[0]),int(coord[1])]=val

    #plt.imshow(Imgrot, cmap='gray'
    return Imgrot
```

```python
plt.figure(figsize=(5,5))
for i,deg in zip(range(4),[-20,-10,10,20]):
    plt.subplot(2,2,i+1)
    plt.imshow(rotateImg(-X_test[0],deg),cmap='gray')
    plt.xticks(())
    plt.yticks(())
plt.show()
```

```python
length= len(missclassified_indices)
i=0
while length>=i*i:
    i=i+1
plt.figure(figsize=(20,20))
plt.title("Missclassifications")
for j,ms in zip(range(0,66),missclassified_indices):
    plt.subplot(i,i,j+1)
    plt.imshow(-X_test[ms][:,:,0],cmap='gray')
    plt.xlabel("R:"+str(real[ms])+" P:"+str(predicted[ms]))
    plt.xticks(())
    plt.yticks(())
plt.show()
```
